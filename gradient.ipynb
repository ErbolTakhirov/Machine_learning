```python
import numpy as np
import matplotlib.pyplot as plt

x = 2
y_true = 5


w = 1.0
b = 0.0
learning_rate = 0.1
steps = 10
w_values = []
b_values = []
loss_values = []

for step in range(steps):
    y_pred = w * x + b
    loss = (y_pred - y_true)**2
    dw = 2*(y_pred - y_true)*x
    db = 2*(y_pred - y_true)
    w_values.append(w)
    b_values.append(b)
    loss_values.append(loss)
    w -= learning_rate * dw
    b -= learning_rate * db
    print(f'Шаг {step + 1}: w = {w:.4f}, b = {b:.4f}, loss = {loss:.4f}')
plt.plot(range(1, steps + 1), loss_values, marker='o')
plt.title('Уменьшение ошибки')

plt.xlabel('Шаг')
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```
